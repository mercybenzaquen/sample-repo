```python
#we had to create a new list since histograms need floats and we had strings (unknown)

age2 = []
for point in df['age']:
    if point != 'Unknown':
        age2.append(float(point))
    else:
        age2.append(0)

df['age_2'] = age2

df['age_2'].hist()

df['age'].sort_values()
        
#we still have unknown values plotted
```
